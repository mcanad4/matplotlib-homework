```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
## Combine the data into a single dataset
# Merge two dataframes using an inner join
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
merge_df
```

```python
## Checking the number of mice.
# Grouping the DataFrame by "Mouse ID"
mouse_group = merge_df.groupby("Mouse ID")

# Count how many purchases each player had and create DataFrame
mouse_count = len(pd.DataFrame(mouse_group["Mouse ID"].count()))
mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

mouse_df = merge_df.groupby(['Mouse ID'])['Timepoint'].max().reset_index()
```

```python
## Create a clean DataFrame by dropping the duplicate mouse by its ID.
# merge in volume for that mouse, regimen, and timepoint
clean_df = pd.merge(mouse_df, merge_df, on=["Mouse ID","Timepoint"], how="left")
```

```python
# Checking the number of mice in the clean DataFrame.
clean_count = clean_df.count()
print(clean_count)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Assemble the resulting series into a single summary dataframe.
```

```python
# get mean for each regiment
mean_df = merge_df.groupby(["Drug Regimen"])['Tumor Volume (mm3)'].mean().reset_index()
```

```python
# get median for each regiment
median_df = merge_df.groupby(["Drug Regimen"])['Tumor Volume (mm3)'].median().reset_index()
```

```python
# filter for and then get variance and standard dev for each regiment
capomulin = merge_df.loc[merge_df["Drug Regimen"] == "Capomulin"]

#calculate variance
cap_var_numpy = np.var(capomulin["Tumor Volume (mm3)"],ddof = 0)

#calculate stdev
cap_sd_numpy = np.std(capomulin["Tumor Volume (mm3)"],ddof = 0)

#define variable for drug name
drug = "Capomulin"

#create start of df with drug inf0
cap_data = [[drug,cap_var_numpy,cap_sd_numpy]]
cap_df = pd.DataFrame(cap_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )
```

```python
# repeat process from above for each drug
ceftamin = merge_df.loc[merge_df["Drug Regimen"] == "Ceftamin"]
cef_var_numpy = np.var(ceftamin["Tumor Volume (mm3)"],ddof = 0)
cef_sd_numpy = np.std(ceftamin["Tumor Volume (mm3)"],ddof = 0)
drug = "Ceftamin"
cef_data = [[drug,cef_var_numpy,cef_sd_numpy]]
cef_df = pd.DataFrame(cef_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )

infubinol = merge_df.loc[merge_df["Drug Regimen"] == "Infubinol"]
inf_var_numpy = np.var(infubinol["Tumor Volume (mm3)"],ddof = 0)
inf_sd_numpy = np.std(infubinol["Tumor Volume (mm3)"],ddof = 0)
drug = "Infubinol"
inf_data = [[drug,inf_var_numpy,inf_sd_numpy]]
inf_df = pd.DataFrame(inf_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )

ketapril = merge_df.loc[merge_df["Drug Regimen"] == "Ketapril"]
ket_var_numpy = np.var(ketapril["Tumor Volume (mm3)"],ddof = 0)
ket_sd_numpy = np.std(ketapril["Tumor Volume (mm3)"],ddof = 0)
drug = "Ketapril"
ket_data = [[drug,ket_var_numpy,ket_sd_numpy]]
ket_df = pd.DataFrame(ket_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )

naftisol = merge_df.loc[merge_df["Drug Regimen"] == "Naftisol"]
naf_var_numpy = np.var(naftisol["Tumor Volume (mm3)"],ddof = 0)
naf_sd_numpy = np.std(naftisol["Tumor Volume (mm3)"],ddof = 0)
drug = "Naftisol"
naf_data = [[drug,naf_var_numpy,naf_sd_numpy]]
naf_df = pd.DataFrame(naf_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )

Placebo = merge_df.loc[merge_df["Drug Regimen"] == "Placebo"]
pla_var_numpy = np.var(Placebo["Tumor Volume (mm3)"],ddof = 0)
pla_sd_numpy = np.std(Placebo["Tumor Volume (mm3)"],ddof = 0)
drug = "Placebo"
pla_data = [[drug,pla_var_numpy,pla_sd_numpy]]
pla_df = pd.DataFrame(pla_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )

Propriva = merge_df.loc[merge_df["Drug Regimen"] == "Propriva"]
pro_var_numpy = np.var(Propriva["Tumor Volume (mm3)"],ddof = 0)
pro_sd_numpy = np.std(Propriva["Tumor Volume (mm3)"],ddof = 0)
drug = "Propriva"
pro_data = [[drug,pro_var_numpy,pro_sd_numpy]]
pro_df = pd.DataFrame(pro_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )

Ramicane = merge_df.loc[merge_df["Drug Regimen"] == "Ramicane"]
ram_var_numpy = np.var(Ramicane["Tumor Volume (mm3)"],ddof = 0)
ram_sd_numpy = np.std(Ramicane["Tumor Volume (mm3)"],ddof = 0)
drug = "Ramicane"
ram_data = [[drug,ram_var_numpy,ram_sd_numpy]]
ram_df = pd.DataFrame(pro_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )

Stelasyn = merge_df.loc[merge_df["Drug Regimen"] == "Stelasyn"]
ste_var_numpy = np.var(Stelasyn["Tumor Volume (mm3)"],ddof = 0)
ste_sd_numpy = np.std(Stelasyn["Tumor Volume (mm3)"],ddof = 0)
drug = "Stelasyn"
ste_data = [[drug,ste_var_numpy,ste_sd_numpy]]
ste_df = pd.DataFrame(ste_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )


Zoniferol = merge_df.loc[merge_df["Drug Regimen"] == "Zoniferol"]
zon_var_numpy = np.var(Zoniferol["Tumor Volume (mm3)"],ddof = 0)
zon_sd_numpy = np.std(Zoniferol["Tumor Volume (mm3)"],ddof = 0)
drug = "Zoniferol"
zon_data = [[drug,zon_var_numpy,zon_sd_numpy]]
zon_df = pd.DataFrame(zon_data, 
                      columns = ['Drug Regimen','Variance','Standard Deviation']
                     )
```

```python
#append new rows into 1 dataframe
new_df = cap_df.append(cef_df)
new_df = new_df.append(inf_df)
new_df = new_df.append(ket_df)
new_df = new_df.append(naf_df)
new_df = new_df.append(pla_df)
new_df = new_df.append(pro_df)
new_df = new_df.append(ram_df)
new_df = new_df.append(ste_df)
new_df = new_df.append(zon_df)
new_df.reset_index()
new_df.head()
```

```python
# Merge two dataframes using an inner join
addmean_df = pd.merge(new_df, mean_df, on="Drug Regimen")
finalstats_df = pd.merge(addmean_df, median_df, on="Drug Regimen")
finalstats_df = finalstats_df.rename(columns={"Tumor Volume (mm3)_x": "Mean","Tumor Volume (mm3)_y": "Median"})
finalstats_df.head()
```
